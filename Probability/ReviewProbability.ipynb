Tóm tắt về Xác suất và Ứng dụng:

1. Định nghĩa:
- Xác suất là đại lượng đo lường sự chắc chắn của một sự kiện
- Giá trị từ 0 đến 1, càng gần 1 càng chắc chắn xảy ra

2. Đặc điểm:
- Linh hoạt trong đo lường khả năng xảy ra
- Phản ánh mức độ không chắc chắn của sự kiện
- Giúp đưa ra quyết định trong điều kiện không chắc chắn

3. Ứng dụng thực tế:
- Thương mại điện tử: Dự đoán khả năng mua hàng, hệ thống recommendation
- Cá cược/Bài bạc: Tính toán xác suất thắng/thua
- Bảo hiểm: Đánh giá rủi ro, định giá bảo hiểm
- Nông nghiệp: Dự báo thời tiết, điều tiết mùa màng
- Machine Learning: Phân loại và dự đoán

4. Các ngành học liên quan:
- Xác suất thống kê
- Định giá tài sản tài chính
- Định giá bảo hiểm
- Machine Learning và AI

5. Vai trò trong học máy:
- Nền tảng cho các mô hình phân loại
- Giúp ước lượng và dự đoán kết quả
- Đánh giá độ tin cậy của mô hình

→ Xác suất là công cụ quan trọng trong việc ra quyết định và dự đoán trong nhiều lĩnh vực, đặc biệt trong machine learning và AI.

## 1.1KHÔNG GIAN MẪU (Sample Space)

1. Định nghĩa:
- Ký hiệu: S
- Là tập hợp tất cả các khả năng có thể xảy ra của một sự kiện
- Là nền tảng để xác định xác suất

2. Ví dụ minh họa:
a) Gieo xúc sắc 6 mặt:
   S = {1, 2, 3, 4, 5, 6}

b) Tung đồng xu 2 mặt:
   S = {sấp, ngửa}

3. Cách biểu diễn xác suất:
- P(X = i) hoặc P(i): xác suất của một sự kiện i
- P(1 ≤ X ≤ 4): xác suất của các sự kiện từ 1 đến 4
- X: biến ngẫu nhiên (random variable)

4. Ý nghĩa:
- Giúp xác định rõ phạm vi của các sự kiện có thể xảy ra
- Là cơ sở để tính toán xác suất
- Làm nền tảng cho việc phân tích xác suất thống kê

→ Không gian mẫu là khái niệm cơ bản và quan trọng trong lý thuyết xác suất, giúp định nghĩa và tính toán xác suất của các sự kiện.

## 1.2BIẾN NGẪU NHIÊN (Random Variable)

1. Định nghĩa:
- Là biến có giá trị phụ thuộc vào sự kiện ngẫu nhiên
- Còn gọi là: random variable, aleatory variable, stochastic variable
- Xuất phát từ xác suất thống kê

2. Phân loại:
a) Biến ngẫu nhiên rời rạc:
   - Có tập hợp giá trị hữu hạn, đếm được
   - Ví dụ: Kết quả xúc sắc {1,2,3,4,5,6}
   - Sử dụng phép tổng trong tính toán

b) Biến ngẫu nhiên liên tục:
   - Có giá trị trong một khoảng liên tục
   - Ví dụ: Cân nặng (52-53kg)
   - Sử dụng tích phân trong tính toán

3. Ví dụ thực tế:
- Tung đồng xu/xúc sắc
- Kết quả xổ số
- Cân nặng của người
- Nhiệt độ môi trường

4. Đặc điểm quan trọng:
- Giá trị không xác định trước
- Phụ thuộc vào sự kiện ngẫu nhiên
- Có thể rời rạc hoặc liên tục
- Cách tính toán khác nhau cho mỗi loại

→ Biến ngẫu nhiên là công cụ quan trọng trong xác suất thống kê, giúp mô hình hóa các hiện tượng không chắc chắn trong thực tế.
